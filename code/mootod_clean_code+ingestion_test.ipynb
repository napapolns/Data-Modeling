```python
import csv


def display_data(file_path):
    def drop_columns(file_path, columns_to_drop):
        # Read the existing data from the CSV file
        with open(file_path, 'r') as csv_file:
            # Create a CSV reader object
            csv_reader = csv.reader(csv_file)

            # Get all rows of data
            all_data = list(csv_reader)

        # Find the indices of the columns to drop
        header_row = all_data[0]
        indices_to_drop = []
        for column in columns_to_drop:
            if column in header_row:
                index = header_row.index(column)
                indices_to_drop.append(index)

        # Remove the columns from each row
        for row in all_data:
            for index in sorted(indices_to_drop, reverse=True):
                del row[index]

        # Write the updated data back to the CSV file
        with open(file_path, 'w', newline='') as csv_file:
            # Create a CSV writer object
            csv_writer = csv.writer(csv_file)

            # Write the modified data to the CSV file
            csv_writer.writerows(all_data)

    # Open the CSV file
    with open(file_path, 'r') as csv_file:
        # Create a CSV reader object
        csv_reader = csv.reader(csv_file)

        # Read the headers
        headers = next(csv_reader)

        # Calculate the maximum width for each column
        column_widths = [len(header) for header in headers]
        for row in csv_reader:
            for i, value in enumerate(row):
                column_widths[i] = max(column_widths[i], len(value))

        # Print the headers
        header_line = '| Index | ' + ' | '.join(header.ljust(column_widths[i]) for i, header in enumerate(headers)) + ' |'
        separator_line = '|-------' + '+'.join('-' * width for width in column_widths) + '|'
        print(header_line)
        print(separator_line)

        # Print each row of data with index
        csv_file.seek(0)  # Reset the file pointer to read the data again
        next(csv_file)  # Skip the headers
        for index, row in enumerate(csv_reader, start=1):
            formatted_row = [' ' + value.ljust(column_widths[i]) + ' ' for i, value in enumerate(row)]
            index_cell = ' ' + str(index).ljust(6) + ' '
            print('|' + index_cell + '|' + '|'.join(formatted_row) + '|')

    return drop_columns

from datetime import datetime

def add_data(file_path):
    # Read the existing data from the CSV file
    with open(file_path, 'r') as csv_file:
        # Create a CSV reader object
        csv_reader = csv.reader(csv_file)

        # Get all rows of data
        all_data = list(csv_reader)

    # Get the headers
    headers = all_data[0]

    # Prompt the user to enter new data for each column
    new_data = []
    for header in headers:
        if header == 'time':
            value = input(f"Enter data for {header} (in DD/MM/YYYY HH:MM format): ")
            while True:
                try:
                    # Try to parse the input value as a datetime object
                    datetime.strptime(value, "%d/%m/%Y %H:%M")
                    break  # Break the loop if the input is in the correct format
                except ValueError:
                    print("Invalid input. Please enter a date and time in the format DD/MM/YYYY HH:MM.")
                    value = input(f"Enter data for {header} (in DD/MM/YYYY HH:MM format): ")
        else:
            value = input(f"Enter data for {header}: ")
            while not all(char.isdigit() or char in ['/', ' '] for char in value):
                print("Invalid input. Please enter only numbers, forward slash (/), and space.")
                value = input(f"Enter data for {header}: ")
        new_data.append(value)

    # Write the new data to the CSV file
    with open(file_path, 'a', newline='') as csv_file:
        # Create a CSV writer object
        csv_writer = csv.writer(csv_file)

        # Write the new data as a row to the CSV file
        csv_writer.writerow(new_data)

    print("Data added successfully.")

from datetime import datetime

def correct_data(file_path, row_index, corrected_data=None):
    if corrected_data is None:
        # Read the existing data from the CSV file
        with open(file_path, 'r') as csv_file:
            # Create a CSV reader object
            csv_reader = csv.reader(csv_file)

            # Get all rows of data
            all_data = list(csv_reader)

        # Check if the row index is within the range of available rows
        if row_index >= 0 and row_index < len(all_data):
            # Get the headers
            headers = all_data[0]

            # Prompt the user to enter corrected data for each column
            corrected_data = []
            for header in headers:
                if header == 'time':
                    value = input(f"Enter corrected data for {header} (in DD/MM/YYYY HH:MM format): ")
                    while True:
                        try:
                            # Try to parse the input value as a datetime object
                            datetime.strptime(value, "%d/%m/%Y %H:%M")
                            break  # Break the loop if the input is in the correct format
                        except ValueError:
                            print("Invalid input. Please enter a date and time in the format DD/MM/YYYY HH:MM.")
                            value = input(f"Enter corrected data for {header} (in DD/MM/YYYY HH:MM format): ")
                else:
                    value = input(f"Enter corrected data for {header}: ")
                    while not all(char.isdigit() or char in ['/', ' '] for char in value):
                        print("Invalid input. Please enter only numbers, forward slash (/), and space.")
                        value = input(f"Enter corrected data for {header}: ")
                corrected_data.append(value)
        else:
            print("Invalid row index.")
            return

    # Read the existing data from the CSV file
    with open(file_path, 'r') as csv_file:
        # Create a CSV reader object
        csv_reader = csv.reader(csv_file)

        # Get all rows of data
        all_data = list(csv_reader)

    # Update the row with corrected data
    all_data[row_index] = corrected_data

    # Write the updated data back to the CSV file
    with open(file_path, 'w', newline='') as csv_file:
        # Create a CSV writer object
        csv_writer = csv.writer(csv_file)

        # Write the corrected data to the CSV file
        csv_writer.writerows(all_data)

    print("Data corrected successfully.")

def delete_data(file_path, row_index):
    # Read the existing data from the CSV file
    with open(file_path, 'r') as csv_file:
        # Create a CSV reader object
        csv_reader = csv.reader(csv_file)

        # Get all rows of data
        all_data = list(csv_reader)

    # Check if the row index is within the range of available rows
    if row_index >= 0 and row_index < len(all_data):
        # Remove the row from the data list
        deleted_row = all_data.pop(row_index)

        # Write the updated data back to the CSV file
        with open(file_path, 'w', newline='') as csv_file:
            # Create a CSV writer object
            csv_writer = csv.writer(csv_file)

            # Write the modified data to the CSV file
            csv_writer.writerows(all_data)

        print(f"Deleted row: {deleted_row}")
        print("Data deleted successfully.")
    else:
        print("Invalid row index.")

def main():
    # File path of the CSV file
    file_path = r'/content/drive/MyDrive/madt/cleaned_datatable.csv'

    while True:
        print("\nCustomer Tool Platform")
        print("1. Display Data")
        print("2. Add Data")
        print("3. Correct Data")
        print("4. Delete Data")
        print("5. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            columns_to_drop = ['vendor', 'tip', 'payment']
            drop_columns_func = display_data(file_path)
            drop_columns_func(file_path, columns_to_drop)
        elif choice == '2':
            add_data(file_path)
            display_data(file_path)
        elif choice == '3':
            row_index = int(input("Enter the row index to correct: "))
            correct_data(file_path, row_index)
            display_data(file_path)
        elif choice == '4':
            row_index = int(input("Enter the row index to delete: "))
            delete_data(file_path, row_index)
            display_data(file_path)
        elif choice == '5':
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
